```python
import pandas as pd

data = pd.read_csv('data.csv')
```

```python
data
```

```python
# Number of male and female candidates applied. In total.
# https://stackoverflow.com/questions/28236305/how-do-i-sum-values-in-a-column-that-match-a-given-condition-using-pandas
gender_group = data.groupby('Gender')['Freq'].sum()
```

```python
gender_group
```

```python
# using loc
print("Male: ", data.loc[data['Gender'] == 'Male', 'Freq'].sum())
print("Female: ", data.loc[data['Gender'] == 'Female', 'Freq'].sum())
```

```python
# how many rows of male and how many rows of female.
data['Gender'].value_counts()
```

```python
# Find male and female rejected
gender_admit_group = data.groupby(['Gender', 'Admit'])['Freq'].sum()
```

```python
gender_admit_group
```

```python
# percentage of women admitted from applications
print("Women admitted percentage among women applicants: ",
      (gender_admit_group['Female']['Admitted'] / gender_group['Female']) *100)
print("Men admitted percentage among men applicants: ",
      (gender_admit_group['Male']['Admitted'] / gender_group['Male']) *100)
```

```python
# number of applicants by gender per Department
gender_dept = data.groupby(['Gender', 'Dept'])['Freq'].sum()
gender_dept
```

```python
data_total = data.join(gender_dept, on=['Gender', 'Dept'], rsuffix='_total')
data_total.head(7)
```

```python
data_total['Percentage'] = data_total['Freq']/data_total['Freq_total']
data_total.head(7)
```

```python
data_total.loc[data_total['Admit'] == 'Admitted', ['Percentage', 'Gender', 'Dept']]
```
